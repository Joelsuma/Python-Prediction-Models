```python
#library import
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#y = 2x + 3
x = np.arange(-5.0, 5.0, 0.1)
```

```python
##You can adjust the slope and intercept to verify the changes in the graph
y = 2*(x) + 3
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
#plt.figure(figsize=(8,6))
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

```python
#y = ax^3 + bx^2 + cx + d
x = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph
y = 1*(x**3) + 1*(x**2) + 1*x + 3
y_noise = 20 * np.random.normal(size=x.size)
ydata = y + y_noise
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

```python
#Y = X**2
x = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph

y = np.power(x,2)
y_noise = 2 * np.random.normal(size=x.size)
ydata = y + y_noise
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r')
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

```python
#Y = a + bc**X
X = np.arange(-5.0, 5.0, 0.1)

##You can adjust the slope and intercept to verify the changes in the graph

Y = np.exp(X)
Y_noise = 10 * np.random.normal(size=X.size)
Ydata = Y + Y_noise

plt.plot(X,Y)
plt.plot(X, Ydata,  'bo')
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

```python
#Y = log(X)
X = np.arange(0.1, 5.0, 0.1)

Y = np.log(X)
Ynoise = 0.3 * np.random.normal(size=X.size)
Ydata = Y + Ynoise

plt.plot(X,Y)
plt.plot(X,Ydata, 'bo')
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

```python
##Sigmoidal Y = a + b/(1 + c**(X-d))
X = np.arange(-5.0, 5.0, 0.1)

Y = 1-4/(1+np.power(3, X-2))
Ynoise = 0.4 * np.random.normal(size = X.size)
Ydata = Y + Ynoise

plt.plot(X,Y)
plt.plot(X, Ydata, 'ro')
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```
